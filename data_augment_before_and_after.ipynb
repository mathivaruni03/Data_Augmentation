```python
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np

# Load Fashion-MNIST dataset
(x_train, y_train), (_, _) = tf.keras.datasets.fashion_mnist.load_data()

# Expand dimensions to match image format requirements
x_train = np.expand_dims(x_train, axis=-1).astype("float32") / 255.0

# Define data augmentation
data_augmentation = tf.keras.Sequential([
    tf.keras.layers.RandomFlip('horizontal'),
    tf.keras.layers.RandomRotation(0.1)
])

# Function to display images
def show_images(original, augmented):
    fig, axes = plt.subplots(2, 5, figsize=(10, 4))
    for i in range(5):
        axes[0, i].imshow(original[i].squeeze(), cmap='gray')
        axes[0, i].axis('on')
        axes[1, i].imshow(augmented[i].squeeze(), cmap='gray')
        axes[1, i].axis('on')
    axes[0, 0].set_title('Original Images')
    axes[1, 0].set_title('Augmented Images')
    plt.show()
  

# Select sample images
sample_images = x_train[:5]
augmented_images = data_augmentation(sample_images)

# Convert tensor to numpy array
augmented_images = augmented_images.numpy()

# Display images before and after augmentation
show_images(sample_images, augmented_images)
```
